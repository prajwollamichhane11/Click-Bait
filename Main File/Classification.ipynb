```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import spacy
import en_core_web_sm
```

```python
data = pd.read_csv("dataset.csv")
clickbaits = pd.read_csv("Clickbaits.csv")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Click on the link and get a chance to win $10000
inputtxt = [input()]
# count_vectorizer = CountVectorizer(stop_words='english')
count_vectorizer = CountVectorizer()
giventextCount = count_vectorizer.fit_transform(inputtxt)
```

```python
nlp = en_core_web_sm.load()
```

```python
giventext = "Click on the link and get a chance to win $10000"
testtext = "This is just a check to see the 10000$ count"

doc1 = nlp(giventext)
doc2 = nlp(testtext)
print(doc1.similarity(doc2))
```

```python
def returnSimilarity(fromDataset,fromWebpage):
    doc1 = nlp(fromDataset)
    doc2 = nlp(fromWebpage)
    
    return doc1.similarity(doc2)    
```

```python
df = pd.DataFrame()
df["similarity"] = clickbaits["Clickbaits"][:4000]
print(len(df))
# df["similarity"]
for i in range(4000):
    df["similarity"][i] = returnSimilarity(giventext,df["similarity"][i])
```

```python
if df["similarity"].mean() > 0.70:
    print("ClickBait")
elif df["similarity"].mean() < 0.50:
    print("Not a Clickbait")
else:
    print("Can not Identify!!")
```
