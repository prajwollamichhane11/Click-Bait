```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import spacy
import en_core_web_sm
```

```python
data = pd.read_csv("dataset.csv")
clickbaits = pd.read_csv("Clickbaits.csv")
```

```python
nlp = en_core_web_sm.load()
```

```python
giventext = "Click on the link and get a chance to win $10000"
testtext = "This is just a check to see the 10000$ count"

doc1 = nlp(giventext)
doc2 = nlp(testtext)
print(doc1.similarity(doc2))
```

```python
doc1
```

```python
def returnSimilarity(fromDataset,fromWebpage):
    doc1 = nlp(fromDataset)
    doc2 = nlp(fromWebpage)
    
    return doc1.similarity(doc2)    
```

```python
random_subset = clickbaits.sample(n=3)
random_subset.head()
```

```python
df = pd.DataFrame()


randomSelectedrow = clickbaits.sample(n=4000)
# randomSelectedrow.head()
df["similarity"] = randomSelectedrow["Clickbaits"]
df["similarity"].head()
df = df.reset_index(drop=True)
df.head()
for i in range(4000):
    df["similarity"][i] = returnSimilarity(giventext,df["similarity"][i])
```

```python
if df["similarity"].mean() > 0.70:
    print("ClickBait")
elif df["similarity"].mean() < 0.50:
    print("Not a Clickbait")
else:
    print("Can not Identify!!")
```
