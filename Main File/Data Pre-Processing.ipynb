```python
import pandas as pd
import numpy as np
```

```python
clickbait = pd.read_fwf("clickbait_data", header=None, names=["Clickbaits"])
nonclickbait = pd.read_fwf("non_clickbait_data", header=None, names=["Clickbaits",1,2,3,4])
```

```python
nonclickbait.drop([1,2,3,4],axis=1,inplace=True)
nonclickbait.dropna(inplace=True)
```

```python
nonclickbait.head()

nonclickbait["bait"] = np.zeros((len(nonclickbait),1),dtype=int)
```

```python
nonclickbait
```

```python
clickbait.dropna(inplace=True)
clickbait.head()
```

```python
clickbait.index
clickbait["bait"] = np.ones((len(clickbait),1),dtype=int)
```

```python
clickbait
```

```python
all_dfs = [clickbait,nonclickbait]

for df in all_dfs:
    df.columns = ["Clickbaits","bait"]
data = pd.concat(all_dfs).reset_index(drop=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
data.to_csv()
```
